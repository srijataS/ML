```python
import pandas as pd
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/BERT/dataset/google-play-app-reviews.csv")
df.head(5)
```

```python
data=df[['reviewId','content']]
data
```

```python
from sklearn.model_selection import train_test_split

data_train, data_test = train_test_split(data, test_size=0.999, random_state = 3)
sentence = data_train['content'].to_list()
#sentence = data_train.to_list()
data_train
```

gtr-t5-large

```python
model = SentenceTransformer('sentence-transformers/gtr-t5-large')
```

```python
aspects = ['User Interface Design', 'Navigation', 'Accessibility',
           'Customization', 'Functionality', 'Performance',
           'Responsiveness', 'Security', 'Privacy', 'Permissions',
           'Data Collection', 'Data Sharing', 'Updates', 'Customer support',
           'Reviews and ratings', 'Developer', 'Price',
           'In-app purchases', 'Advertisement Frequency', 'Battery/power Drainage']
```

```python
data=[]
for i , id in zip(sentence, data_train['reviewId']):
  embeddings= model.encode(i, convert_to_tensor=True)
  row = []
  for aspect in aspects:
    aspect_embedding =  model.encode(aspect, convert_to_tensor=True)
      # We use cosine-similarity and torch.topk to find the highest 5 scores
    cos_scores = util.cos_sim(embeddings, aspect_embedding)[0][0].item()
    row.append(cos_scores)
  data.append(row)
print(data)
df = pd.DataFrame(data, index=zip(data_train['reviewId'],sentence), columns=aspects)
df.head(5)
```

```python
print(len(data),len(aspects),len(sentence))
```
